```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import network
import vis
import patternLib
from copy import deepcopy
```

This file is walkthrough of a demo case showcasing my implementation of a Hopfield Network model

```python
pattern_size = 5

# create the hopfield network object with 25 neurons
hnet = network.myHopfieldNetwork(nuerons=pattern_size**2)
```

```python
# create a list of 5 random patterns
plist = patternLib.get_patterns(pattern_size,5)

# plot the random patterns
vis.see_pattern_list(plist)
```

```python
# calculate the weights from the pattern list
wgts = hnet.calculate_weights(plist)
```

```python
cue = deepcopy(plist)
pattern_to_disturb = 1        # select between 0 and pattern_size - 1
num_changes = 5               # number of neurons to distort in selected pattern

# Create a the distorted pattern
S = patternLib.perturb_pattern(pattern=cue[pattern_to_disturb], num_changes=num_changes)

# plot of the noisy pattern
print(f"We have modified pattern P{pattern_to_disturb} with a change in {num_changes} neurons\n")
vis.see_pattern(S)
```

```python
# let the states evolve using the hopfield network (you can use "sync" or "async" update)
state_list = hnet.network_evolution(S, wgts, method="sync", steps = 4)

# plot the sequence of state evolution along with overlap
vis.plot_state_sequence_and_overlap(state_list,plist,ref_idx=pattern_to_disturb,suptitle=f"Network Dynamics for a modified P{pattern_to_disturb}")
```
